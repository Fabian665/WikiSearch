```python
import pickle
from math import log10
```

```python
def normalize_page_views(page_views):
      """
      This function normalizes a dictionary of page views using min-max scaling.
    
      Args:
          page_views: A dictionary of document IDs (keys) and their corresponding page views (values).
    
      Returns:
          A dictionary with the same keys as the input, but with the page views normalized between 0 and 1.
      """
    
      # Find the minimum and maximum page views
      min_views = min(page_views.values())
      max_views = max(page_views.values())
    
      # Normalize the page views
      normalized_views = {}
      for doc_id, page_view in page_views.items():
          normalized_views[doc_id] = (page_view - min_views) / (max_views - min_views)
    
      return normalized_views
```

```python
normalized_page_views = {did: log10(views + 1) for did, views in page_views.items()}
```
